```python
#Converting the values object and boolean to integers
admissions['boarding'] = admissions['boarding'].replace({True: 1, False: 0})
admissions['faculty_child'] = admissions['faculty_child'].replace({True: 1, False: 0})
admissions['legacy_child'] = admissions['legacy_child'].replace({True: 1, False: 0})

admissions.to_csv("/content/admission_updated.csv", index=False)

```

```python
data = pd.read_csv('admission_updated.csv')
```

```python
data['gender'] = data['gender'].astype(str)
data['ethnicity'] = data['ethnicity'].astype(str)
```

```python
data['gender'] = data['gender'].fillna('Unknown')
data['ethnicity'] = data['ethnicity'].fillna('Unknown')

#Converting columns to str
data['gender'] = data['gender'].astype(str)
data['ethnicity'] = data['ethnicity'].astype(str)

print(data.dtypes)
```

```python

data['gender'] = data['gender'].apply(str)
data['ethnicity'] = data['ethnicity'].apply(str)

print(data.dtypes)
```

```python
data['gender'] = data['gender'].apply(lambda x: str(x) if x is not None else 'Unknown')
data['ethnicity'] = data['ethnicity'].apply(lambda x: str(x) if x is not None else 'Unknown')
```
